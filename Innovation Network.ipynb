```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("AIML Proposals.xlsx", sheet_name="Edge list")
df=df.rename(index=str, columns={"Idea Team": "Idea Team1"})
```

```python
df=df.loc[:, "Submitter.1":"Idea Team10"]
df['id'] = df.index
df.head()
```

```python
temp=pd.wide_to_long(df, ["Idea Team"], i="id", j="num")
temp2=temp.dropna()
temp.head()
```

```python
%matplotlib notebook

g=nx.from_pandas_edgelist(temp, source='Submitter.1', target='Idea Team',)
nx.draw(g, with_labels=True, font_size=5)
plt.show()
```

```python
%matplotlib notebook
g=nx.from_pandas_edgelist(temp, source='Submitter.1', target='Idea Team')
g2=g.to_undirected()
nx.draw(g2, with_labels=True, node_color='y', font_size=5)
plt.show()
```

```python
%matplotlib notebook

g=nx.from_pandas_edgelist(temp2, source='Submitter.1', target='Idea Team',)
g2=g.to_undirected()
nx.draw(g2, with_labels=True, node_size=7, font_size=5)
plt.show()
```
